```python
import pandas as pd
```

#### Data Preparation and Data Cleaning component

```python
df1 = pd.read_csv("Monthly Maximum Temperature.csv") 
df1.head()
df2 = pd.read_csv("Monthly Minimum Temperature.csv") 
df2.head()
```

```python
df1 = df1.loc[df1["District"] == "Medchal-Malkajgiri"]
df1 = df1.loc[df1["Mandal"] == "Kapra"]
df1.head()
```

```python
df2 = df2.loc[df2["District"] == "Medchal-Malkajgiri"]
df2 = df2.loc[df2["Mandal"] == "Kapra"]
df2.head()
```

```python
frames = [df1,df2]
df3 = pd.concat(frames)
df3 = df3.reset_index(drop=True)
df3.head()
```

```python
df3 = df3.T
df3.head()
```

```python
df3 = df3.reset_index()
df3
```

```python
df3 = df3.drop([0,1,2])
df3 = df3.reset_index(drop = True)
df3.head()
```

```python
df3.index.name = 'n_in'
df3 = df3.rename(columns = {"index":"Date",0:"Max Temp",1:"Min Temp"})
print(df3.info())
```

```python
df3["Date"]=pd.to_datetime(df3["Date"],format="%b-%y")
df3.head()
```

#### Data Visualization component

```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'iframe' # or 'colab' or 'iframe' or 'iframe_connected' or 'sphinx_gallery'
```

```python
fig = px.line(df3, x="Date", y=["Max Temp","Min Temp"], title='Max/Min Temperatures of Kapra Mandal from July 2013 to Jan 2018', 
              labels = {"Date":"Months"})
fig.update_layout(yaxis_title="Temperature in °Celcius")
fig.show()
```
