```python
from owid import catalog
import pandas as pd
```

#### Data acquisition

```python
url = "https://ourworldindata.org/the-worlds-deadliest-earthquakes"
table = pd.read_html(url)
df1 = table[0]
df1
```

```python
table = catalog.find('natural_disasters')
df2 = table.iloc[1].load()
df2
```

### Data Preparation and Data Cleaning

```python
#choose only those rows that correspond to earthquakes and are for years 2020 or higher.
df2.reset_index()
cond = (df2.index.get_level_values(2) == 'earthquake') & (df2.index.get_level_values(1)>=2000)
df2 = df2[cond]
df2
```

```python
#showing all the rows that need to be removed
df2[df2.index.get_level_values(0).isin(['World',
                            'European Union (27)',
                            'Africa', 
                            'Antarctica', 
                            'Asia', 'Europe',
                            'North America', 
                            'Oceania', 
                            'South America'
                            'Upper-middle-income countries', 
                            'Lower-middle-income countries', 
                            'High-income countries', 
                            'Low-income countries'])]
```

```python
df3 = df2[df2.index.get_level_values(0).isin(['World',
                            'European Union (27)',
                            'Africa', 
                            'Antarctica', 
                            'Asia', 'Europe',
                            'North America', 
                            'Oceania', 
                            'South America'     
                            'Upper-middle-income countries', 
                            'Lower-middle-income countries', 
                            'High-income countries', 
                            'Low-income countries']) == False]
df3
```

```python
#showing that the rows have been removed
df3[df3.index.get_level_values(0).isin(['World',
                            'European Union (27)',
                            'Africa', 
                            'Antarctica', 
                            'Asia', 'Europe',
                            'North America', 
                            'Oceania', 
                            'South America'
                            'Upper-middle-income countries', 
                            'Lower-middle-income countries', 
                            'High-income countries', 
                            'Low-income countries'])]
```
