```python
from owid import catalog
import pandas as pd
```

#### Data acquisition

```python
url = "https://ourworldindata.org/the-worlds-deadliest-earthquakes"
table = pd.read_html(url)
df1 = table[0]
df1
```

```python
table = catalog.find('natural_disasters')
df2 = table.iloc[1].load()
df2
```

### Data Preparation and Data Cleaning

```python
df2 = df2.reset_index()
```

```python
#choose only those rows that correspond to earthquakes and are for years 2020 or higher.
df2 = df2[df2["year"]>=2000]
df2 = df2[df2["type"]=="earthquake"]
df2
```

```python
#showing all the rows that need to be removed
rows_to_drop = ["World", "European Union (27)", "Upper-middle-income countries", "Asia", "Africa", "North America", "South America", "Europe", "Australia", "Antarctica", "Lower-middle-income countries",
                 "High-income countries", "Low-income countries"]
```

```python
df3 = df2[~df2["country"].isin(rows_to_drop)]
df3
```

```python
#showing that the rows have been removed
df3 = df3.sort_values(by = "total_dead", ascending = False).head(5)
df3
```

```python
df3 = df3[['country','year','total_dead']]
df3
```

```python
df3['dfname'] = 'Deadly since 2000'
```

```python
df1.rename(columns = {'Location':'country', 'Year':'year', 'Estimated death toll':'total_dead'}, inplace = True)
df1 = df1[['country','year','total_dead']]
df1
```

```python
df1['country'] = df1['country'].str.split(pat=", ").str[1]
df1['dfname'] = "10 Deadliest Quakes in History"
df1
```

```python
combined = pd.concat([df1,df3],axis=0)
combined.reset_index(inplace=True)
combined = combined.drop(columns = ["index"])
combined
```

```python
combined = combined.drop_duplicates(subset = ["country", "year"])
combined.reset_index(inplace = True)
combined = combined.drop(columns = ["index"])
combined = combined.sort_values(by = 'total_dead',ascending = False)
combined
```

### Plotting the data

```python
import plotly.express as px
```

```python
plot_dat = combined
plot_dat['country'] = combined['country'] + ' ' + combined['year'].astype(int).astype(str)
plot_dat = plot_dat.sort_values(by = 'total_dead',ascending=True)
plot_dat
```

```python
plot_dat.sort_values(by = 'year',ascending=False)
```

```python
plot_dat = plot_dat.rename(columns = {"dfname":"Killer Quakes"})
```

```python
px.bar(plot_dat,x='total_dead',y='country',text='total_dead',color='Killer Quakes',title = "Killer Quakes")
```
