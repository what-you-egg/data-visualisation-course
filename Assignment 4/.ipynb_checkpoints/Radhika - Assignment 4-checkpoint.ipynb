```python
import nsepython
import pandas as pd
import plotly.express as px
```

```python
start_date = "01-Mar-2011"
end_date = "31-Aug-2023"

nifty50_data = index_total_returns("NIFTY 50", start_date, end_date)
niftybank_data = index_total_returns("NIFTY BANK", start_date, end_date)
niftyit_data = index_total_returns("NIFTY IT", start_date, end_date)
```

```python
nifty50_data['TotalReturnsIndex'] = nifty50_data['TotalReturnsIndex'].astype(float)
niftybank_data['TotalReturnsIndex'] = niftybank_data['TotalReturnsIndex'].astype(float)
niftyit_data['TotalReturnsIndex'] = niftyit_data['TotalReturnsIndex'].astype(float)
```

```python
nifty50_data['NIFTY 50 Change'] = ((nifty50_data['TotalReturnsIndex'] - nifty50_data['TotalReturnsIndex'].iloc[-1]) / nifty50_data['TotalReturnsIndex'].iloc[-1]) * 100
niftybank_data['NIFTY BANK Change'] = ((niftybank_data['TotalReturnsIndex'] - niftybank_data['TotalReturnsIndex'].iloc[-1]) / niftybank_data['TotalReturnsIndex'].iloc[-1]) * 100
niftyit_data['NIFTY IT Change'] = ((niftyit_data['TotalReturnsIndex'] - niftyit_data['TotalReturnsIndex'].iloc[-1]) / niftyit_data['TotalReturnsIndex'].iloc[-1]) * 100
```

```python
combined_data = pd.concat([nifty50_data['Date'], nifty50_data['NIFTY 50 Change'], niftybank_data['NIFTY BANK Change'], niftyit_data['NIFTY IT Change']], axis=1)
combined_data = combined_data.dropna()
```

```python
combined_data
```

```python
fig = px.line(combined_data, x ="Date", y=["NIFTY 50 Change", "NIFTY BANK Change", "NIFTY IT Change"], title='Percentage Change in Index funds Since May 2014', width=800, height=400)
fig.update_layout(
    xaxis_title='Date',
    yaxis_title='value'
)

fig.show()
```
